````python
#Write your code here
!wget https://raw.githubusercontent.com/venkatareddykonasani/Datasets/master/IPC/THE_INDIAN_PENAL_CODE.pdf

from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("THE_INDIAN_PENAL_CODE.pdf")
Pages = loader.load()

# Extract text from the PDF:

#Loop through the pages of the PDF document and concatenate the
# text content of each page into a single string.
docs=""
for page in Pages:
  docs += page.page_content

#Print the first 1000 characters of the extracted text:
print(docs[:1000])
````

````python
# Print the number of lines, words, and characters in the extracted docs:

# Use string methods to split the docs by newline characters to count the number of lines.
# Use string methods to split the docs by spaces to count the number of words.
# Use the len function to count the total number of characters in the docs.

docs_lines = docs.split("\n")
docs_words = docs.split(" ")
docs_chars = len(docs)

print("Number of lines:", len(docs_lines))
print("Number of words:", len(docs_words))
print("Number of characters:", docs_chars)

````

````python
#Write your code here
from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.chains import ConversationalRetrievalChain
````

````python
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
retriever = IPC_db.as_retriever()

conversational_RAG = ConversationalRetrievalChain.from_llm(llm= llm,
                                              retriever= retriever,
                                              memory= memory,
                                            )
````
