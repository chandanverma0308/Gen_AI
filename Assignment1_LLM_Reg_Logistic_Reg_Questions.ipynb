```python
from sklearn.linear_model import LogisticRegression
!wget https://raw.githubusercontent.com/venkatareddykonasani/Datasets/master/Credit%20Risk%20Data_balanced/Credit_risk_data_bal_v2.csv
```

```python
import pandas as pd
df= pd.read_csv("Credit_risk_data_bal_v2.csv")
df.head()
```

## Q3.1) Write code to import data from the below location

https://raw.githubusercontent.com/venkatareddykonasani/Datasets/master/Credit%20Risk%20Data_balanced/Credit_risk_data_bal_v2.csv

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['SeriousDlqin2yrs']), df['SeriousDlqin2yrs'], test_size=0.2, random_state=42 )

```

## Q3.2)  Write code for Splitting the data into Train Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['SeriousDlqin2yrs']), df['SeriousDlqin2yrs'], test_size=0.2, random_state=42 )

```

## Q3.3) Write code to build the model. Use 'SeriousDlqin2yrs' as target  variable and rest all as predictor variables. Print the beta co-efficients

```python
####Model training

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

## Q3.4)Perform Model Validation and Cross Validation on train and test data. Print the confusion matrix and accuracies

```python
## Model validation
# get the prediction of both data

y_train_pred= model.predict(X_train)
y_test_pred=  model.predict(X_test)


#confusion matrix
from sklearn.metrics import confusion_matrix
cm_training= confusion_matrix(y_train,y_train_pred )
cm_testing = confusion_matrix(y_test,y_test_pred)

print("Train Confusion Matrix:\n", cm_training)
print("\nTest Confusion Matrix:\n", cm_testing)



```

```python
from sklearn.metrics import accuracy_score

## accuracy
trianing_accuracy = accuracy_score(y_train,y_train_pred)
Tetsing_accuracy = accuracy_score(y_test,y_test_pred)

print("Train Accuracy:", trianing_accuracy)
print("Test Accuracy:", Tetsing_accuracy)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    model,
    df.drop(columns=['SeriousDlqin2yrs']),
    df['SeriousDlqin2yrs'],
    cv=5,              # 5-fold CV
    scoring="accuracy"
)

print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", cv_scores.mean())
```
